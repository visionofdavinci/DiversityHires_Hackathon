```python
from src.calendar_agent import authenticate, get_all_busy_events
from datetime import datetime, timedelta
from src.utils.time_utils import merge_events, find_free_slots

service = authenticate()

busy_events = get_all_busy_events(service)
merged_busy = merge_events(busy_events)

start = datetime.now()
end = start + timedelta(days=7)

free_slots = find_free_slots(merged_busy, start, end, min_duration_minutes=30)

for start, end in free_slots:
    print(f"Free from {start} to {end}")
```
